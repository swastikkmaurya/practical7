```python
#Factorial of a Number
n= int(input('Enter a number : '))
fact = 1
while n>0:
   fact *=n
   n-=1
print("Factorial = ",fact)
```

```python
#Write a python code fibonacci series
n_terms = int(input("How many terms of the Fibonacci series do you want? "))

if n_terms <= 0:
        print("Please enter a positive integer.")
else:
        a, b = 0, 1
        count = 0

        print(f"Fibonacci series up to {n_terms} terms:")

        while count <= n_terms:
            print(a,end=" ")
            nth = a + b
            a = b
            b = nth
            count += 1
        print()
```

```python
#Reverse A number
num = int(input("Enter a number : "))
rev=0
while num > 0:
  digit = num % 10
  rev = rev * 10 + digit
  num //= 10
print("Reversed Number ",rev)
```

```python
# Check Palindrome Number
num = int(input("Enter a number : "))
temp = num
rev=0
while num > 0:
  digit = num % 10
  rev = rev * 10 + digit
  num //= 10
if temp == rev:
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
#
```

```python
#check palindrome for string
string = input("Enter a string : ")
rev = string[::-1]
if string==rev:
    print("Palindrome")
else:
    print("Palindrome Not ")
```

```python
s = "madam"
i,j = 0,len(s) - 1
is_palindrome = True
while i < j:
  if s[i]!=s[j]:
    is_palindrome = False
  break
i+=1
j-=1
if is_palindrome:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

```python
#Count Digit in a number
num = int(input("Enter a number : "))
count = 0

while num > 0:
  count += 1
  num //= 10
print("Number of digits : ",count)
```

```python
#Search an element in a list
nums = [10,20,30,40,50]
key = int(input("Enter a number : "))
i = 0
while i < len(nums):
   if nums [i] == key:
    print("Element Found at Index ",i)
    break
   i +=1
else:
  print("Not found")
```
